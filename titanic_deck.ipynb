```python
import seaborn as sns
titanic_df= sns.load_dataset("titanic")
titanic_df
```

```python
sns.countplot(x="deck",data=titanic_df)
```

```python
# sibsp（配偶者と兄弟）とparch（子と両親）列で、単身かそうでないかを判定
alone_s=(titanic_df['sibsp']+titanic_df['parch'])
alone_s
```

```python
titanic_df.loc[alone_s > 0, 'with'] = 'True'
titanic_df.loc[alone_s == 0, 'with'] = 'False'
３titanic_df
```

```python
# 同伴者がいるかいないか、席ごとにカウント
sns.countplot(x='with',hue = 'pclass', data=titanic_df)
```

```python
sns.factorplot(x="pclass",y="survived",hue="sex",data=titanic_df)
```
