```python
# DataSeriesの中身を確認して条件分岐しようとしたが、DataSeriesはIf文で検証はできない。
def make_passenger_culumn(titanic_df):
    if titanic_df['Age']<15:
        titanic_df['Passenger']='Child' 
    elif titanic_df['Age'].isnull:
        titanic_df['Passenger']=Nan
    return titanic_df['Passenger']
```

```python
# 値を参照して、新しい列を作るのはloc関数を使う。
def make_passenger_culumn(titanic_df):
    titanic_df.loc[titanic_df['Age'] <= 15, 'Passenger'] = 'child'
    titanic_df.loc[titanic_df['Age'] > 15, 'Passenger'] = titanic_df['Sex']
    titanic_df.loc[titanic_df['Age'].isnull(), 'Passenger'] = titanic_df['Sex']
    return titanic_df
```

```python
ps_titanic_df= make_passenger_culumn(titanic_df)
```
